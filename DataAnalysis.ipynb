```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Error handling for data loading
try:
    # Load Iris dataset
    iris = load_iris()
    df = pd.DataFrame(iris.data, columns=iris.feature_names)
    df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
    print("Data loaded successfully.\n")
except Exception as e:
    print(f"Error loading dataset: {e}")
```

```python
# Task 1: Basic Statistics
print("Basic Statistics using .describe():")
print(df.describe())
```

```python
# Task 2: Grouping 
grouped = df.groupby('species').mean()
print("\nMean of numerical columns grouped by species:")
print(grouped)

```

```python
# Optional Insights
print("\nðŸ“Œ Insights:")
print("- Versicolor flowers have the longest average petal length.")
print("- Setosa species generally have the smallest features overall.")
```

```python
# Set Seaborn theme
sns.set(style="whitegrid")
```

```python
# 1. Line Chart - Petal Length over index (trending view)
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x=df.index, y='petal length (cm)', hue='species')
plt.title("Trend of Petal Length Across Samples")
plt.xlabel("Sample Index")
plt.ylabel("Petal Length (cm)")
plt.legend(title='Species')
plt.show()
```

```python
# 2. Bar Chart - Average Petal Length per Species
plt.figure(figsize=(8, 6))
sns.barplot(x='species', y='petal length (cm)', data=df)
plt.title("Average Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Average Petal Length (cm)")
plt.show()
```

```python
# 3. Histogram - Distribution of Sepal Length
plt.figure(figsize=(8, 6))
sns.histplot(df['sepal length (cm)'], bins=20, kde=True)
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# 4. Scatter Plot - Sepal Length vs Petal Length
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal length (cm)', y='petal length (cm)', hue='species')
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title='Species')
plt.show()
```
